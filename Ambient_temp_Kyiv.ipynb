```python
#Importing the required libraries
import pandas as pd 
import requests
from bs4 import BeautifulSoup

#Downloading contents of the web page
webpage_response = requests.get('https://www.climatestotravel.com/climate/ukraine/kiev')

#Creating a BeautifulSoup object
soup = BeautifulSoup(webpage_response.content, 'html.parser')
```

```python
#Scraping a table with the climate data Using Attributes
table = soup.find(attrs={'class':'cities'})
```

```python
#Defining the dataframe
df = pd.DataFrame(columns=["Month", "Min(°C)", "Max(°C)", "Mean(°C)"])

#Collecting data from the table
for row in table.find_all("tr"):    

    columns = row.find_all("td")
    if(columns != []):
        min_C = columns[1].get_text()
        max_C = columns[2].get_text()
        mean_C = columns[3].get_text()

    df = df.append({ "Min(°C)": min_C, "Max(°C)": max_C, 'Mean(°C)': mean_C}, ignore_index=True)


print(df)
```
