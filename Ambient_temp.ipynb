```python
#Importing the required libraries
import pandas as pd
import re

#Parsing HTML elements into a lis of data frames
df = pd.read_html("https://en.wikipedia.org/wiki/List_of_cities_by_average_temperature")
```

```python
#Adding an appropriate region to each country based on a table
df[0]["Region"] = "Africa"
df[1]["Region"] = "Asia"
df[2]["Region"] = "Europe"
df[3]["Region"] = "North America"
df[4]["Region"] = "Oceania"
df[5]["Region"] = "South America"
```

```python
#Concatenating all datafrmaes into a single one
full_df = pd.concat([df[0], df[1], df[2], df[3], df[4], df[5]])

#Removing unnecessary columns and placing Region at the beginning 
new_df = full_df[["Region", "Country", "City", "Jan", "Feb",
                         "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]]
#Calling a function
new_df
```

```python
#Tydying the data to have Celcius only, without parentheses
def strip(x):
    return re.sub("[\(\[].*?[\)\]]", "", x)
new_df["Jan"] = new_df.apply(lambda col: strip, axis=0)
```
